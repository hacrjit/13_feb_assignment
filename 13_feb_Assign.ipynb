### Question No. 2

```python
import inspect
print("The class hierarchy for built-in exceptions is:")
inspect.getclasstree(inspect.getmro(Exception))
def classtree(cls, indent=0):
    print('.' * indent, cls.__name__)
    for subcls in cls.__subclasses__():
        classtree(subcls, indent + 3)
classtree(Exception)
```

### Question No. 3

The ArithmeticError class in Python is a base class for a number of built-in exceptions that are raised when an arithmetic error occurs.<br> These errors include:<br>
<b>ZeroDivisionError:</b> This error is raised when you try to divide by zero.<br> For example, the following code will raise a ZeroDivisionError:
10 / 0<br>
<b>OverflowError:</b> This error is raised when the result of an arithmetic operation is too large to be represented in the target data type.<br> For example, the following code will raise an OverflowError:
1 << 100


### Question No. 4

<b>LookupError</b>: The base class for the exceptions that are raised when a key or index used on a mapping or sequence is invalid.<br>
<b>KeyError</b>:
Raised when a mapping (dictionary) key is not found in the set of existing keys.<br>
<b>IndexError</b>: 
Raised when a sequence subscript is out of range. (Slice indices are silently truncated to fall in the allowed range; if an index is not an integer, TypeError is raised.)
